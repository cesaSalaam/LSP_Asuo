```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.graphics.mosaicplot import mosaic
```

# **Heat Maps**

```python
# Some sector and geographies for illustration
sector=["Technology","Consumer Discretionary", "Real Estate", "Financial Services"]

geography=["Ghana", "Nigeria", "South Africa", "Haiti"]



# Simulate market returns based on a random 
returns= np.array([[40,13,20,37],
         [30,41,37,46],
         [89,50,65,55],
         [35,20,25,35]])

fig, ax = plt.subplots()
im = ax.imshow(returns)

# We want to show all ticks...
ax.set_xticks(np.arange(len(sector)))
ax.set_yticks(np.arange(len(geography)))

# ... and label them with the respective list entries
ax.set_xticklabels(sector)
ax.set_yticklabels(geography)


ax.set_title("Country revenue by sector in billions")
fig.tight_layout()
plt.show()
```

# **Correlograms**

We usually use a correlogram to display correlation data over a time period.

```python
# data Import 
# Fixing random state for reproducibility
np.random.seed(2434548631)
x, y = np.random.randn(2, 100)

fig, ax1  = plt.subplots()

ax1.xcorr(x, y, usevlines=True, maxlags=50, normed=True, lw=2)

plt.show()
```
